```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.random.rand(50)
```

```python
y=np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x,y,c='#223BDc',alpha=.9,marker='X')
plt.xlabel("this is my x axis")
plt.ylabel('this is my y axis')
plt.title('this is my x vs y')
plt.grid()
```

```python
x=np.linspace(1,10,100)
```

```python
y=np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(8,2))
plt.plot(x,y,'--r')
plt.xlabel('this is my x axis')
plt.ylabel('this is my y label')
plt.title('this is my sin wave graph')
plt.grid()
```

```python
x=['a','b','c','d','e']
y=np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(8,4))
plt.bar(x,y,color='#7E10D5')
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.title('this is my bar plot')
plt.grid()
```

```python
plt.figure(figsize=(8,4))
plt.barh(x,y,color='#7E10D5')
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.title('this is my bar plot')
plt.grid()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z,c='#9327E8',alpha=.9)
plt.xlabel('this is my x axis')
plt.ylabel('this is my y axis')
plt.show()
```
